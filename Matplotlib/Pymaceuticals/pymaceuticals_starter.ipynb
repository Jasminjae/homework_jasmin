# Matplotlib Homework

```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_data = pd.read_csv(mouse_drug_data_to_load)
clinicaltrial_data = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
combined_data = pd.merge(clinicaltrial_data, mouse_data, how='left')


# Display the data table for preview
combined_data.head(10)
```

#### Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint
modified_data = combined_data.groupby(['Drug', 'Timepoint']).mean()[f'Tumor Volume (mm3)']
#Convert to DataFrame
mean_data = pd.DataFrame(modified_data)
# Preview DataFrame
mean_data2 = mean_data.reset_index()
mean_data2.head(10)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
sem_modified_data = combined_data.groupby(['Drug', 'Timepoint']).sem()[f'Tumor Volume (mm3)']
# Convert to DataFrame
sem_data = pd.DataFrame(sem_modified_data)
#locked_df = sorted_data.loc[:, ['Drug', 'Timepoint']]
sem_data2 = sem_data.reset_index()
# Preview DataFrame
sem_data2.head(10)
```

```python
# Minor Data Munging to Re-Format the Data Frames
flipped_mean_df = mean_data2.pivot(index='Timepoint', columns='Drug')[f'Tumor Volume (mm3)']
flipped_sem_df = sem_data2.pivot(index='Timepoint', columns='Drug')[f'Tumor Volume (mm3)']
# Preview that Reformatting worked
flipped_mean_df.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(flipped_mean_df.index, flipped_mean_df['Capomulin'], 
             yerr=flipped_sem_df['Capomulin'], color='red', 
             marker='o', markersize=3, linestyle='dashed', linewidth=1)
plt.errorbar(flipped_mean_df.index, flipped_mean_df['Infubinol'], 
             yerr=flipped_sem_df['Infubinol'], color='blue', 
             marker='x', markersize=3, linestyle='dashed', linewidth=1)
plt.errorbar(flipped_mean_df.index, flipped_mean_df['Ketapril'], 
             yerr=flipped_sem_df['Ketapril'], color='green', 
             marker='*', markersize=3, linestyle='dashed', linewidth=1)
plt.errorbar(flipped_mean_df.index, flipped_mean_df['Placebo'], 
             yerr=flipped_sem_df['Placebo'], color='purple', 
             marker='^', markersize=3, linestyle='dashed', linewidth=1)

plt.title('Tumor: Response to Treatment')
plt.xlabel('Time in Days')
plt.ylabel(f'Tumor Volume (mm3)')
plt.grid(axis='y')
plt.legend(['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'], fontsize='small')
# Save the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
met_mean_data = combined_data.groupby(['Drug', 'Timepoint']).mean()['Metastatic Sites']
# Convert to DataFrame
met_mean_df = pd.DataFrame(met_mean_data)
# Preview DataFrame
met_mean_df = met_mean_df.reset_index()
met_mean_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
met_sem_data = combined_data.groupby(['Drug', 'Timepoint']).sem()['Metastatic Sites']
# Convert to DataFrame
met_sem_df = pd.DataFrame(met_sem_data)
# Preview DataFrame
met_sem_df = met_sem_df.reset_index()
met_sem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_meta_flipped = met_mean_df.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
sem_meta_flipped = met_sem_df.pivot(index='Timepoint', columns='Drug', values='Metastatic Sites')
# Preview that Reformatting 
mean_meta_flipped.head()
```

```python
# Generate the Plot (with Error Bars)
plt.errorbar(mean_meta_flipped.index, mean_meta_flipped['Capomulin'], 
             yerr=sem_meta_flipped['Capomulin'], color='yellow', 
             marker='o', markersize=3, linestyle='solid', linewidth=1)
plt.errorbar(mean_meta_flipped.index, mean_meta_flipped['Infubinol'], 
             yerr=sem_meta_flipped['Infubinol'], color='lime', 
             marker='o', markersize=3, linestyle='solid', linewidth=1)
plt.errorbar(mean_meta_flipped.index, mean_meta_flipped['Ketapril'], 
             yerr=sem_meta_flipped['Ketapril'], color='orange', 
             marker='o', markersize=3, linestyle='solid', linewidth=1)
plt.errorbar(mean_meta_flipped.index, mean_meta_flipped['Placebo'], 
             yerr=sem_meta_flipped['Placebo'], color='black', 
             marker='o', markersize=3, linestyle='solid', linewidth=1)

plt.title('Metastatic Sites: Response to Treatment')
plt.xlabel('Time in Days')
plt.ylabel('Metastatic Sites')
plt.grid(axis='y')
plt.legend(['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'], fontsize='small')
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mice_count_df = combined_data.groupby(['Drug', 'Timepoint']).count()['Mouse ID']

# Convert to DataFrame
mice_count_df= pd.DataFrame(mice_count_df)
# Preview DataFrame
survival_df = mice_count_df.reset_index()
survival_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
survival_df_flipped = survival_df.pivot(index='Timepoint', columns='Drug', values='Mouse ID')
# Preview that Reformatting 
survival_df_flipped.head()
```

```python
# Generate the Plot (Accounting for percentages)

plt.plot(survival_df_flipped['Capomulin'], color='blue', marker='x', linestyle='dashed')
plt.plot(survival_df_flipped['Infubinol'], color='red', marker='o', linestyle='dashed')
plt.plot(survival_df_flipped['Ketapril'], color='grey', marker='*', linestyle='dashed')
plt.plot(survival_df_flipped['Placebo'], color='aqua', marker='^', linestyle='dashed')
# Save the Figure
plt.title('Survival Rate of Mice')
plt.xlabel('Time in Days')
plt.ylabel('Survival Rate')
plt.grid(axis='both')
plt.legend(['Capomulin', 'Infubinol', 'Ketapril', 'Placebo'], fontsize='small')
# Show the Figure
plt.show()
```
